```python
import yfinance as yahooFinance
import datetime
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
start = "2020-01-01"
end = '2024-09-08'
```

```python
NVDA = yahooFinance.download('NVDA', start=start, end=end)['Adj Close']
SPX = yahooFinance.download('^GSPC', start=start, end=end)['Adj Close']
MSFT = yahooFinance.download('MSFT', start=start, end=end)['Adj Close']
AAPL = yahooFinance.download('AAPL', start=start, end=end)['Adj Close']
```

```python
NVDA_returns = (NVDA / NVDA.iloc[0] - 1) * 100
SPX_returns = (SPX / SPX.iloc[0] - 1) * 100
MSFT_returns = (MSFT / MSFT.iloc[0] - 1) * 100
AAPL_returns = (AAPL / AAPL.iloc[0] - 1) * 100
```

```python
plt.figure(figsize=(10,6))
plt.plot(NVDA_returns, label='NVDA')
plt.plot(SPX_returns, label='S&P 500')
plt.plot(MSFT_returns, label='MSFT')
plt.plot(AAPL_returns, label='AAPL')

plt.text(NVDA_returns.index[-1], NVDA_returns[-1], f'{NVDA_returns[-1]:.2f}%', color='blue', ha='left', va='center')
plt.text(SPX_returns.index[-1], SPX_returns[-1] - 2, f'{SPX_returns[-1]:.2f}%', color='orange', ha='left', va='center')
plt.text(MSFT_returns.index[-1], MSFT_returns[-1] - 0, f'{MSFT_returns[-1]:.2f}%', color='green', ha='left', va='center')
plt.text(AAPL_returns.index[-1], AAPL_returns[-1] + 12, f'{AAPL_returns[-1]:.2f}%', color='red', ha='left', va='center')


plt.title('Returns of NVDA, S&P 500, MSFT, and AAPL')
plt.xlabel('Date')
plt.ylabel('Returns')
plt.legend()
plt.grid(True)
plt.show()
```
